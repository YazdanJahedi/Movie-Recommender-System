```python
import numpy as np
import pandas as pd
```

```python
movies = pd.read_csv('./IMDB/movies_metadata.csv', low_memory=False)
movies.head()
```

# -----------------------------
# Feature Engineering
# -----------------------------


```python
movies.shape
```

## adult column:

```python
print('audult unique values:', movies['adult'].nunique())
print('audult values:', movies['adult'].unique())
```

```python
# True or False values for adult is valid. other rows should be deleted:
movies = movies[(movies['adult'] == 'True') | (movies['adult'] == 'False')]
movies.shape
```

```python
# Replace True to 1 and False to 0 in the column:
movies.replace({'adult': {'True': 1, 'False': 0}}, inplace=True)
movies.head() 
```

## homepage, imdb_id, overview and post_path and tagline

```python
# These columns are not informative in this case. So we should delete them.
movies.drop(['homepage', 'imdb_id', 'overview', 'poster_path', 'tagline'], axis=1, inplace=True)
```

## belongs_to_collection
assing 1 to those which have collection and 0 to others

Actually we don't use specific collection name. Just using has_collection value. This is for generalization!

```python
movies['has_collection'] = movies['belongs_to_collection'].apply(lambda x: 0 if pd.isna(x) else 1)
```

```python
movies.drop('belongs_to_collection', axis=1, inplace=True)
movies.head()
```

## budget  and  revenue
First we Should change dtype to int.

Next, it has missing values. So we should replace them.

(?) mean is a good replacement?

(?) normalize or standarize?

```python
movies["budget"] = pd.to_numeric(movies["budget"])
movies["revenue"] = pd.to_numeric(movies["revenue"])
movies.dtypes
```

```python
# Replace 0 values to mean of all values:
b_mean = movies["budget"].mean()
r_mean = movies["revenue"].mean()
movies.replace({'budget': {0: b_mean}, 'revenue':{0: r_mean}}, inplace=True)
movies.head()
```

```python
# standardize these 2 cols
movies["std_budget"] = (movies["budget"] - movies["budget"].mean()) / movies["budget"].std()
movies["std_revenue"] = (movies["revenue"] - movies["revenue"].mean()) / movies["revenue"].std()
movies.drop(['revenue', 'budget'], axis=1, inplace=True)
movies.head()
```

## title and original_title
We can save one of this columns. saving both is not necessary

```python
movies.drop('original_title', axis=1, inplace=True)
```

## spoken_languages and original_language
We should use one of these columns to generalize the modle. using both is not informative enough.

To simplify, we can choose original_langue and delete spoken_language

In the next step, we should use one-hot endcoder

```python
movies.drop('spoken_languages', axis=1, inplace=True)
```

```python
encoded = pd.get_dummies(movies['original_language'], prefix='lang')
movies = pd.concat([movies, encoded], axis=1).drop('original_language', axis=1)
movies.head()
```

## production_companies* and production_countries
To generalize the model, we can delete these two cols.

We can check production_companies in model later...

```python
movies.drop(['production_companies','production_countries'], axis=1 , inplace=True)
```

## popularity

set proper dtype and standadization

```python
movies["popularity"] = pd.to_numeric(movies["popularity"])
```

```python
movies.popularity.max()
```

```python
movies['std_popularity'] = (movies.popularity - movies.popularity.mean()) / movies.popularity.std()
movies.drop('popularity', axis=1, inplace =True)
movies.head()
```

## release_date

binning with decade of the movies: lables are ``

```python
movies["release_year"] = pd.to_datetime(movies["release_date"], errors='coerce').dt.year
```

```python
bins = [1800, 1900, 1910, 1920, 1930, 1940, 1950, 1960, 1970, 1980, 1990, 2000, 2010, 2020, 2030]
labels = ["1800's", "1900's", "1910's", "1920's", "1930's", "1940's", "1950's", "1960's", "1970's", 
          "1980's", "1990's", "2000's", "2010's", "2020's"]
movies['bin_year'] = pd.cut(movies['release_year'], bins=bins, labels=labels)
```

```python
encoded = pd.get_dummies(movies['bin_year'], prefix='released_in')
movies = pd.concat([movies, encoded], axis=1)
movies.drop(['release_date', 'release_year', 'bin_year'], axis=1, inplace=True)
movies.head()
```

## runtime

binning this col to three groups: `short_movie`, `normal_movies`, `long_movie` and `series`

```python
bins = [0, 45, 120, 250, 99999999999]
labels = ['short_movie', 'normal_movies', 'long_movie', 'series']
movies['bin_runtime'] = pd.cut(movies['runtime'], bins=bins, labels=labels)
```

## status

```python
movies['status'].unique()
# we can select just Released ones...
```

```python
movies = movies[movies.status == 'Released' ]
movies.shape
```

## video
Replace True with 1 and False with 0

```python
movies.replace({'video': {False: 0, True: 1}}, inplace=True)
movies.head()
```


## genres
First we should explode all diferent genres for a movie and make a one-hot matrix from it. Then concat one-hot to the main dataFram.

```python
from ast import literal_eval
```

```python
movies['genres'] = movies['genres'].apply(literal_eval)
```

```python
encoded = pd.get_dummies(movies.explode('genres')['genres'].astype('str'), prefix='genre')
encoded = encoded.groupby(level=0).sum()
encoded
```

```python
movies = pd.concat([movies, encoded], axis=1).drop('genres', axis=1)
movies.head()
```

## vote_average and vote_count
normalize vote_average column and standardization on vote_count data

```python
movies['norm_vote_avg'] = (movies.vote_average - movies.vote_average.min()) / (movies.vote_average.max() - movies.vote_average.min())
movies['std_vote_cnt'] = (movies.vote_average - movies.vote_average.min()) / (movies.vote_average.std())
movies.drop(['vote_average', 'vote_count'], axis=1, inplace=True)
movies.head()
```

## id

a few movies doesn't have id. So we can delete them

next we set id as index of the df

```python
print(movies.shape)
print(movies.id.nunique())
```

```python
movies['id'] = pd.to_numeric(movies['id'], errors='coerce')
movies = movies[~movies.id.isna()]
```

```python
movies.set_index('id', inplace=True)
movies.sort_values('id', ascending=True, inplace=True)
```
