```python
def cal_cost(index):
    u = user_mat[index]#.reshape((-1, 1))
    t = user_ratings[index] # target
    y = u.dot(movie_mat)
    diff = y - t
    diff2 = np.power(diff, 2)
    # n = np.argwhere(~np.isnan(u)).size
    n = np.count_nonzero(~np.isnan(t))
    if n == 0:
        return 'Unknown'
    return np.nansum(diff2)/(2 * n)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
for i in range(user_mat.shape[0]):
    # index of all rated values for this user
    rated_indexes = np.argwhere(~np.isnan(user_ratings[i])).T.tolist()[0]
    
    y = user_ratings[i, rated_indexes]
    x = movie_mat[:-1, rated_indexes] # the last one is b and shouldn't be used in w estimation

    lr_model = LinearRegression()
    lr_model.fit(x.T, y)
    w = lr_model.coef_
    b = lr_model.intercept_
    
    user_mat[i] = a = np.append(w, b)
    
    print('for: ', i, '  the cost is:  ', cal_cost(i))
```

```python
def cal_sim(ind1, ind2):
    user1 = user_ratings[ind1]
    user2 = user_ratings[ind2]
    dot = np.nansum(user1 * user2) # dot product ignore missign values
    u1_l = np.sqrt(np.nansum(np.power(user1, 2))) # ||user1||
    u2_l = np.sqrt(np.nansum(np.power(user2, 2))) # ||user2||
    
    if u1_l == 0 or u2_l == 0:
        return 0
        
    return dot / (u1_l * u2_l)
```

```python
def sim_vector(user_index):
    no_users = user_ratings.shape[0]
    sim_vec = np.zeros(no_users)
    for i in range(no_users):
        sim_vec[i] = cal_sim(user_index, i)
        
    return sim_vec
```

```python
def cal_r_s(user_index):
    sim_vec = sim_vector(user_index)
    r_s_vector = sim_vec * user_ratings.T
    return r_s_vector.T, sim_vec
```

```python
def col_filtering():
    no_users = user_ratings.shape[0]
    no_user_index = user_ratings.shape[1]
    new_user_ratings = np.zeros(user_ratings.shape)
    for u_ind in range(no_users):
        r_s, s = cal_r_s(u_ind)
        for i in range(no_user_index):
            not_nan_ind = np.argwhere(~np.isnan(r_s[:,i])).T.tolist()[0]
            r_s_sum = np.nansum(r_s[:,i]) 
            
            s_selected = s[not_nan_ind]
            s_sum = np.nansum(s_selected) - 1 
            
#             if i == 1:
#                 print('soorat:  ', r_s_sum, '    makhraj:  ', s_sum)
                
            val = r_s_sum / s_sum
            new_user_ratings[u_ind, i] = val
            
    print('done...')
    return new_user_ratings
```

```python
def recommended_index(user_ind, items):
    # content base
    u_rat1 = user_mat[user_ind].dot(movie_mat)
    # collaborative base
    u_rate2 = trained_col_filt[user_ind]
    
    u_sum = u_rat1 + u_rate2
    
    least_max = np.sort(u_sum)[-items]
    
    return np.argwhere(u_sum >= least_max).T.tolist()[0]
```
