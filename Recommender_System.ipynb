```python
import numpy as np
import pandas as pd
```

```python
movies = pd.read_csv('./temp/cleaned_movies_metadata.csv')
ratings = pd.read_csv('./IMDB/ratings_small.csv')
```

```python
cond = ~ratings['movieId'].isin(movies['id'])
ratings.drop(ratings[cond].index, inplace=True)
```

```python
movies.set_index('id', inplace=True)
movies.drop('title', axis=1, inplace=True)
ratings.drop('timestamp', axis=1, inplace=True)
```
