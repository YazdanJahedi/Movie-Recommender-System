```python
movie_mat = movies.values.T
movie_mat.shape
```

```python
user_mat = np.random.rand(ratings.userId.nunique(),movie_mat.shape[0])
user_mat.shape
```

# ------------------------------

```python
def cal_cost(index):
    u = user_mat[index]#.reshape((-1, 1))
    t = user_ratings[index] # target
    y = u.dot(movie_mat)
    diff = y - t
    diff2 = np.power(diff, 2)
    # n = np.argwhere(~np.isnan(u)).size
    n = np.count_nonzero(~np.isnan(t))
    if n == 0:
        n = 1
    return np.nansum(diff2)/(2 * n), y, t, n
```

```python
def gradient_descent(index, alpha):
    w = user_mat[index]   
    cost = float("inf")
    cnt = 0
    while True:
        
        min_cost, y, t, n = cal_cost(index)
        
#         print("min cost is   " , min_cost)

        if min_cost > cost or cnt > 50:
            break

        cost = min_cost

        diff = y - t
        print('---->' , diff)
        where_are_NaNs = np.isnan(diff)
        diff[where_are_NaNs] = 0
        
        print(diff)
        deriv = np.dot(movie_mat, diff)
        print('deriv:', deriv)
        w -= ((alpha / n) * deriv)
#         print("---------\n")
        
        cnt +=1
    return w, cost
```

```python
w, c = gradient_descent(0, alpha)
```

```python
alpha = 0.1

for i in range(user_mat.shape[0]):
    w, c = gradient_descent(i, alpha)
    print('for i:', i, '-->    ', c)
```
